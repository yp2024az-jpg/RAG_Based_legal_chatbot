## 2. Legal Data Schema & Validation

```python
class JurisdictionEnum(Enum):
    """Supported jurisdictions"""
    INDIA = "India"
    DELHI = "Delhi"
    MAHARASHTRA = "Maharashtra"
    KARNATAKA = "Karnataka"
    TAMIL_NADU = "Tamil Nadu"

class LegalCategoryEnum(Enum):
    """Legal categories"""
    CRIMINAL_LAW = "Criminal Law"
    CIVIL_LAW = "Civil Law"
    PROCEDURAL_LAW = "Procedural Law"
    CONTRACT_LAW = "Contract Law"
    FAMILY_LAW = "Family Law"
    PROPERTY_LAW = "Property Law"
    LABOR_LAW = "Labor Law"
    TAX_LAW = "Tax Law"
    COMMERCIAL_LAW = "Commercial Law"
    CONSTITUTIONAL_LAW = "Constitutional Law"

class LegalSectionSchema:
    """Schema for legal sections"""
    
    REQUIRED_FIELDS = {'id', 'title', 'jurisdiction', 'category', 'year', 'content'}
    OPTIONAL_FIELDS = {'amendments', 'tags', 'references', 'keywords', 'source', 'updated_at'}
    
    @staticmethod
    def create_section(id: str, title: str, jurisdiction: str, category: str, 
                      year: int, content: str, **kwargs) -> Dict:
        """Create a valid legal section"""
        section = {
            'id': id,
            'title': title,
            'jurisdiction': jurisdiction,
            'category': category,
            'year': year,
            'content': content,
            'created_at': datetime.now().isoformat(),
            'updated_at': datetime.now().isoformat(),
        }
        # Add optional fields
        for key in LegalSectionSchema.OPTIONAL_FIELDS:
            if key in kwargs:
                section[key] = kwargs[key]
        return section
    
    @staticmethod
    def validate_section(section: Dict) -> Tuple[bool, str]:
        """Validate a legal section"""
        missing = LegalSectionSchema.REQUIRED_FIELDS - set(section.keys())
        if missing:
            return False, f"Missing required fields: {missing}"
        return True, "Valid"

print("✓ Legal data schema initialized")
print(f"Required fields: {LegalSectionSchema.REQUIRED_FIELDS}")
print(f"Optional fields: {LegalSectionSchema.OPTIONAL_FIELDS}")
```

```python
# Install required packages for web scraping
required_packages = ['requests', 'beautifulsoup4']
for package in required_packages:
    try:
        __import__(package.replace('-', '_'))
        print(f"✓ {package} already installed")
    except ImportError:
        print(f"Installing {package}...")
        subprocess.check_call([sys.executable, '-m', 'pip', 'install', package, '-q'])
        print(f"✓ {package} installed")

import requests
from bs4 import BeautifulSoup
import time
```

```python
class LegalWebScraper:
    """Web scraper for legal sections from online sources"""
    
    def __init__(self, timeout=10, delay=1):
        self.timeout = timeout
        self.delay = delay
        self.session = requests.Session()
        self.session.headers.update({
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
        })
    
    def scrape_ipc_section(self, section_number: int) -> Optional[Dict]:
        """Scrape IPC section from indiankanoon.org"""
        url = f"https://indiankanoon.org/doc/1350048/"
        try:
            response = self.session.get(url, timeout=self.timeout)
            response.raise_for_status()
            time.sleep(self.delay)
            # Parse and extract section data
            # Note: This is a template - actual parsing depends on website structure
            return {'status': 'success', 'section': section_number}
        except requests.exceptions.RequestException as e:
            logger.error(f"Error scraping section {section_number}: {e}")
            return None
    
    def batch_scrape_sections(self, section_numbers: List[int]) -> List[Dict]:
        """Scrape multiple sections"""
        results = []
        for section_num in section_numbers:
            result = self.scrape_ipc_section(section_num)
            if result:
                results.append(result)
        return results
    
    def validate_scraped_data(self, data: Dict) -> bool:
        """Validate scraped data against schema"""
        return all(field in data for field in LegalSectionSchema.REQUIRED_FIELDS)

print("✓ Web scraper class created")
print("\nNote: In production, scraper would extract from:")
print("  - indiankanoon.org (comprehensive legal database)")
print("  - indiacode.nic.in (official legislation)")
print("  - legislative.gov.in (parliamentary resources)")
```

## 7. Legal Section Manager (CRUD Operations)

```python
class LegalSectionManager:
    """Manager for CRUD operations on legal sections"""
    
    def __init__(self, database_path: Path):
        self.database_path = database_path
        self.sections = self._load_database()
        self.index = {s['id']: i for i, s in enumerate(self.sections)}
    
    def _load_database(self) -> List[Dict]:
        """Load sections from JSON file"""
        if self.database_path.exists():
            with open(self.database_path, 'r', encoding='utf-8') as f:
                return json.load(f)
        return []
    
    def _save_database(self):
        """Save sections to JSON file"""
        with open(self.database_path, 'w', encoding='utf-8') as f:
            json.dump(self.sections, f, indent=2, ensure_ascii=False)
    
    def create_section(self, id: str, title: str, jurisdiction: str, 
                      category: str, year: int, content: str, **kwargs) -> Dict:
        """Create new legal section"""
        if id in self.index:
            raise ValueError(f"Section {id} already exists")
        
        section = LegalSectionSchema.create_section(
            id, title, jurisdiction, category, year, content, **kwargs
        )
        is_valid, msg = LegalSectionSchema.validate_section(section)
        if not is_valid:
            raise ValueError(f"Invalid section: {msg}")
        
        self.sections.append(section)
        self.index[id] = len(self.sections) - 1
        self._save_database()
        logger.info(f"Created section {id}")
        return section
    
    def read_section(self, section_id: str) -> Optional[Dict]:
        """Read legal section by ID"""
        if section_id not in self.index:
            return None
        return self.sections[self.index[section_id]]
    
    def update_section(self, section_id: str, **kwargs) -> Dict:
        """Update legal section"""
        if section_id not in self.index:
            raise ValueError(f"Section {section_id} not found")
        
        idx = self.index[section_id]
        section = self.sections[idx]
        
        # Update allowed fields
        updatable = {'content', 'amendments', 'tags', 'references', 'keywords'}
        for key, value in kwargs.items():
            if key in updatable:
                section[key] = value
        
        section['updated_at'] = datetime.now().isoformat()
        self._save_database()
        logger.info(f"Updated section {section_id}")
        return section
    
    def delete_section(self, section_id: str) -> bool:
        """Delete legal section"""
        if section_id not in self.index:
            return False
        
        idx = self.index[section_id]
        self.sections.pop(idx)
        # Rebuild index
        self.index = {s['id']: i for i, s in enumerate(self.sections)}
        self._save_database()
        logger.info(f"Deleted section {section_id}")
        return True
    
    def search_sections(self, query: str) -> List[Dict]:
        """Search sections by title or content"""
        query_lower = query.lower()
        results = [
            s for s in self.sections
            if query_lower in s['title'].lower() or query_lower in s['content'].lower()
        ]
        return results
    
    def filter_by_category(self, category: str) -> List[Dict]:
        """Filter sections by category"""
        return [s for s in self.sections if s['category'] == category]
    
    def get_statistics(self) -> Dict:
        """Get database statistics"""
        categories = {}
        for section in self.sections:
            cat = section['category']
            categories[cat] = categories.get(cat, 0) + 1
        
        return {
            'total_sections': len(self.sections),
            'categories': categories,
            'date_range': {
                'oldest': min((s['year'] for s in self.sections), default=None),
                'newest': max((s['year'] for s in self.sections), default=None)
            }
        }

print("✓ Legal Section Manager class created")
```

## 9. Integration with RAG Pipeline

```python
# Load embedder for RAG integration
from sentence_transformers import SentenceTransformer

try:
    embedder = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
    print("✓ Embedder loaded successfully")
except:
    print("Installing sentence-transformers...")
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'sentence-transformers', '-q'])
    embedder = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
    print("✓ Embedder installed and loaded")

print(f"Embedding dimension: {embedder.get_sentence_embedding_dimension()}")
```

```python
class LegalRAGIntegration:
    """Integration of legal database with RAG pipeline"""
    
    def __init__(self, manager: LegalSectionManager, embedder):
        self.manager = manager
        self.embedder = embedder
        self.embeddings = {}
        self._generate_embeddings()
    
    def _generate_embeddings(self):
        """Generate embeddings for all sections"""
        for section in self.manager.sections:
            section_id = section['id']
            # Combine title and content for embedding
            text = f"{section['title']}. {section['content'][:500]}"
            embedding = self.embedder.encode(text, convert_to_numpy=True)
            self.embeddings[section_id] = embedding
    
    def semantic_search(self, query: str, top_k: int = 5) -> List[Tuple[Dict, float]]:
        """Semantic search over legal sections"""
        query_embedding = self.embedder.encode(query, convert_to_numpy=True)
        
        scores = []
        for section in self.manager.sections:
            section_id = section['id']
            section_embedding = self.embeddings[section_id]
            # Cosine similarity
            from sklearn.metrics.pairwise import cosine_similarity
            similarity = cosine_similarity(
                query_embedding.reshape(1, -1),
                section_embedding.reshape(1, -1)
            )[0][0]
            scores.append((section, similarity))
        
        # Sort by similarity and return top-k
        scores.sort(key=lambda x: x[1], reverse=True)
        return scores[:top_k]
    
    def hybrid_search(self, query: str, top_k: int = 5) -> List[Tuple[Dict, float]]:
        """Hybrid search combining semantic and keyword search"""
        # Semantic results
        semantic_results = self.semantic_search(query, top_k=top_k*2)
        # Keyword results
        keyword_results = self.manager.search_sections(query)
        
        # Combine and deduplicate
        combined = {}
        for section, score in semantic_results:
            combined[section['id']] = (section, score * 0.6)  # 60% semantic weight
        
        for section in keyword_results[:top_k*2]:
            if section['id'] in combined:
                _, semantic_score = combined[section['id']]
                combined[section['id']] = (section, semantic_score + 0.4)  # Add 40% keyword weight
            else:
                combined[section['id']] = (section, 0.4)
        
        # Sort and return top-k
        results = sorted(combined.values(), key=lambda x: x[1], reverse=True)[:top_k]
        return results

print("✓ Legal RAG Integration class created")
```
